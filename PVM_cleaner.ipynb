## This section crates a new column called "Type", which will be used to show the ratio of towable units vs. motorized units in future visualizations. 

```python
#create type column
towable = ['FW', 'TT', 'PT', 'CT', 'TC', 'CA', 'BT']
motorized = ['A', 'B', 'C', 'AU']
type = []

for row in clean_df['Class']:
    if row in towable:
        type.append('Towable')
    elif row in motorized:
        type.append('Motorized')
    else:
        type.append('Inc Input')
        
clean_df.insert(5,'Type',type,True)

clean_df.sample(5)
```

## Final cleaning, replacing 'NaN's with 0 and removing the comma from the 'Porter ID' column, then download!

```python
# import Numpy
import numpy as np
# drop NaNs
clean_df = clean_df.replace(np.nan,0)
clean_df.sample(10)
```

```python
# remove the comma from Porter ID and change to int
clean_df[["Porter ID"]] = clean_df[["Porter ID"]].replace(',','', regex=True).astype(int)
clean_df.sample(10)
```
